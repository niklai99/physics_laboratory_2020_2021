```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import seaborn as sns
```

```python
simulation = '../Simulations/PreAmp/Data/PreAmp_planning.txt'

data = pd.read_csv(simulation, sep = '\t', index_col = False)
```

```python
# FIG SETTINGS AND AXES
fig = plt.figure(figsize=(16, 16))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# PLOT SIMULATIONS
ax1.plot(data['time'], data['V(in)'], color = '#0e94ff', linewidth = 3, label = '$V_{in}$')
ax1.plot(data['time'], data['V(out)'], color = '#ff790e', linewidth = 3, label = '$V_{out}$')

# DRAW DASHED 'ZERO' LINE
ax1.axhline(color = '#000000', linewidth = 1, linestyle = 'dashed')

# PLOT TITLE
ax1.set_title('PreAmp - Preliminary Simulation', fontsize = 32)

# AXIS LABELS
#ax1.set_xlabel('Time (s)', fontsize = 26, loc = 'center')
ax1.set_ylabel('Signal (V)', fontsize = 26, loc = 'center')

# AXIS TICKS
ax1.tick_params(axis = 'both', which = 'major', labelsize = 22, direction = 'in', length = 10)
ax1.tick_params(axis = 'both', which = 'minor', labelsize = 22, direction = 'in', length = 5)
ax1.set_xticks(ticks = ax1.get_xticks(), minor = True)
ax1.set_yticks(ticks = ax1.get_yticks(), minor = True)
ax1.minorticks_on()
ax1.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax1.xaxis.get_offset_text().set_fontsize(22)
ax1.ticklabel_format(axis = 'x', style = 'scientific', scilimits = (0, 0))

# PLOT RANGE
ax1.set_xlim(left = 0, right = 5e-3)
ax1.set_ylim(bottom = -1.1, top = 0.5)

# MAKE LEGEND
ax1.legend(loc = 'lower right', prop = {'size': 26}, ncol = 1, frameon = True, fancybox = False, framealpha = 1)

# PLOT SIMULATIONS
ax2.plot(data['time'], data['V(in)'], color = '#0e94ff', linewidth = 3, label = '$V_{in}$')
ax2.plot(data['time'], data['V(out)'], color = '#ff790e', linewidth = 3, label = '$V_{out}$')

# DRAW DASHED 'ZERO' LINE
ax2.axhline(color = '#000000', linewidth = 1, linestyle = 'dashed')

# AXIS LABELS
ax2.set_xlabel('Time (s)', fontsize = 26, loc = 'center')
ax2.set_ylabel('Signal (V)', fontsize = 26, loc = 'center')

# AXIS TICKS
ax2.tick_params(axis = 'both', which = 'major', labelsize = 22, direction = 'in', length = 10)
ax2.tick_params(axis = 'both', which = 'minor', labelsize = 22, direction = 'in', length = 5)
ax2.set_xticks(ticks = ax1.get_xticks(), minor = True)
ax2.set_yticks(ticks = ax1.get_yticks(), minor = True)
ax2.minorticks_on()
ax2.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax2.xaxis.get_offset_text().set_fontsize(22)
ax2.ticklabel_format(axis = 'x', style = 'scientific', scilimits = (0, 0))

# PLOT RANGE
ax2.set_xlim(left = 0.975e-3, right = 1.2e-3)
ax2.set_ylim(bottom = -1.1, top = 0.5)

# MAKE LEGEND
ax2.legend(loc = 'lower right', prop = {'size': 26}, ncol = 1, frameon = True, fancybox = False, framealpha = 1)

# SAVE FIGURE
fig.savefig('../Plots/PreAmp_preliminary_simulation.png', dpi = 300)

plt.show()
```

```python
freq = '../Simulations/PreAmp/Data/PreAmp_frequency.txt'

data = pd.read_csv(freq, sep = '\t', index_col = False)

data['f'] = np.log10(data['f'])
```

```python
# FIG SETTINGS AND AXES
fig = plt.figure(figsize=(16, 8))
ax1 = fig.add_subplot(1, 1, 1)

# PLOT SIMULATIONS
ax1.plot(data['f'], data['H'], color = '#0e94ff', linewidth = 3, label = 'H')

# PLOT TITLE
ax1.set_title('PreAmp - Bode Preliminary Simulation', fontsize = 32)

# AXIS LABELS
ax1.set_xlabel('Freq (dec)', fontsize = 26, loc = 'center')
ax1.set_ylabel('H (dB)', fontsize = 26, loc = 'center')

# AXIS TICKS
ax1.tick_params(axis = 'both', which = 'major', labelsize = 22, direction = 'in', length = 10)
ax1.tick_params(axis = 'both', which = 'minor', labelsize = 22, direction = 'in', length = 5)
ax1.set_xticks(ticks = ax1.get_xticks(), minor = True)
ax1.set_yticks(ticks = ax1.get_yticks(), minor = True)
ax1.minorticks_on()
ax1.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax1.xaxis.get_offset_text().set_fontsize(22)
ax1.ticklabel_format(axis = 'x', style = 'scientific', scilimits = (0, 0))

# PLOT RANGE
ax1.set_xlim(left = 1, right =6)
ax1.set_ylim(bottom = -40, top = 25)

# MAKE LEGEND
#ax1.legend(loc = 'lower right', prop = {'size': 26}, ncol = 1, frameon = True, fancybox = False, framealpha = 1)

fig.savefig('../Plots/PreAmp_preliminary_bode_simulation.png', dpi = 300)

plt.show()
```
