```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import seaborn as sns
```

```python
simulation = '../Simulations/PreAmp/Data/PreAmp_planning.txt'

data = pd.read_csv(simulation, sep = '\t', index_col = False)
```

```python
# FIG SETTINGS AND AXES
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(1, 1, 1)

# PLOT SIMULATIONS
ax.plot(data['time'], data['V(in)'], color = '#0e94ff', linewidth = 3, label = '$V_{in}$')
ax.plot(data['time'], data['V(out)'], color = '#ff790e', linewidth = 3, label = '$V_{out}$')

# DRAW DASHED 'ZERO' LINE
ax.axhline(color = '#000000', linewidth = 1, linestyle = 'dashed')

# PLOT TITLE
ax.set_title('PreAmp - Preliminary Simulation', fontsize = 32)

# AXIS LABELS
ax.set_xlabel('Time (s)', fontsize = 26, loc = 'center')
ax.set_ylabel('Signal (V)', fontsize = 26, loc = 'center')

# AXIS TICKS
ax.tick_params(axis = 'both', which = 'major', labelsize = 22, direction = 'in', length = 10)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 22, direction = 'in', length = 5)
ax.set_xticks(ticks = ax.get_xticks(), minor = True)
ax.set_yticks(ticks = ax.get_yticks(), minor = True)
ax.minorticks_on()
ax.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax.xaxis.get_offset_text().set_fontsize(22)
ax.ticklabel_format(axis = 'x', style = 'scientific', scilimits = (0, 0))

# PLOT RANGE
ax.set_xlim(left = 0, right = 5e-3)
ax.set_ylim(bottom = -1.1, top = 0.5)

# MAKE LEGEND
ax.legend(loc = 'lower right', prop = {'size': 26}, ncol = 1, frameon = True, fancybox = False, framealpha = 1)

# SAVE FIGURE
fig.savefig('../Plots/PreAmp_preliminary_simulation.png', dpi = 300)

plt.show()
```
