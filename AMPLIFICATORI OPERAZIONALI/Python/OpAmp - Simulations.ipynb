```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import seaborn as sns
```

```python
simulation1 = '../Data/Simulations/opamp_05v.txt'
simulation2 = '../Data/Simulations/opamp_3v.txt'

data1 = pd.read_csv(simulation1, sep = '\t', index_col = False, header = None)
data2 = pd.read_csv(simulation2, sep = '\t', index_col = False, header = None)

data1.index = np.arange(1, len(data1)+1)
data2.index = np.arange(1, len(data2)+1)

data1.columns = ['t', 'Vin', 'Vout']
data2.columns = ['t', 'Vin', 'Vout']
```

```python
# FIG SETTINGS AND AXES
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(1, 1, 1)

# PLOT SIMULATIONS
ax.plot(data1['t'] * 1e3, data1['Vin'], color = '#0095FF', linewidth = 3, label = '$V_{in}$ @ 0.5V')
ax.plot(data1['t'] * 1e3, data1['Vout'], color = '#0045FF', linewidth = 3, label = '$V_{out}$ @ 0.5V')
ax.plot(data2['t'] * 1e3, data2['Vin'], color = '#FF790E', linewidth = 3, label = '$V_{in}$ @ 3V')
ax.plot(data2['t'] * 1e3, data2['Vout'], color = '#FF1800', linewidth = 3, label = '$V_{out}$ @ 3V')

# DRAW DASHED 'ZERO' LINE
ax.axhline(color = '#000000', linewidth = 1, linestyle = 'dashed')

# AXIS LABELS
ax.set_xlabel('Time (ms)', fontsize = 26)
ax.set_ylabel('Signal (V)', fontsize = 26)

# AXIS TICKS
ax.tick_params(axis = 'both', which = 'major', labelsize = 22, direction = 'in', length = 10)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 22, direction = 'in', length = 5)
ax.set_xticks(ticks = ax.get_xticks(), minor = True)
ax.set_yticks(ticks = ax.get_yticks(), minor = True)
ax.minorticks_on()
#ax.ticklabel_format(axis = 'x', style = 'scientific', scilimits = (0, 0))
ax.xaxis.get_offset_text().set_fontsize(15)
ax.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))

# PLOT RANGE
ax.set_xlim(left = 10, right = 14)
ax.set_ylim(bottom = -17, top = 17)

# MAKE LEGEND
ax.legend(loc = 'lower right', prop = {'size': 26}, ncol = 2, frameon = True, fancybox = False, framealpha = 1)

# SAVE FIGURE
fig.savefig('../Plots/Report_Plots/opamp_spice_py.png', dpi = 300)

plt.show()
```
